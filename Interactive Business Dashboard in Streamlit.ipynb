```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global_Superstore2.csv", encoding="latin1")


total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()

print("ðŸ’° Total Sales:", round(total_sales, 2))
print("ðŸ“ˆ Total Profit:", round(total_profit, 2))

sales_by_cat = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(6,4))
sales_by_cat.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()

profit_by_region = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(6,4))
profit_by_region.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Profit by Region")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()

top_customers = df.groupby("Customer Name")["Sales"].sum().nlargest(5)

plt.figure(figsize=(6,4))
top_customers.plot(kind="bar", color="salmon", edgecolor="black")
plt.title("Top 5 Customers by Sales")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```
